```python
import socket
import sys

# --- CẤU HÌNH ---
HOST = "172.23.8.153"   # IP server PostgreSQL
PORT = 5432             # Port PostgreSQL

def check_port():
    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    sock.settimeout(5)  # Chờ tối đa 5 giây
    try:
        result = sock.connect_ex((HOST, PORT))
        if result == 0:
            print(f"THÀNH CÔNG: Port {PORT} trên {HOST} ĐANG MỞ!")
            print(f"→ PostgreSQL có thể kết nối được!")
            return True
        else:
            print(f"THẤT BẠI: Port {PORT} trên {HOST} BỊ ĐÓNG hoặc BỊ CHẶN!")
            return False
    except Exception as e:
        print(f"LỖI: {e}")
        return False
    finally:
        sock.close()

if __name__ == "__main__":
    check_port()
```
